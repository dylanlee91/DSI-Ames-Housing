```python
train = pd.read_csv('datasets/train.csv')
```

```python
def subplot_scat(data, independentcol, dependentcol, title=None, x_label=None, y_label=None,figuresize = None, columns = 2):
    rows = int(round((len(dependentcol)/columns),0)) # Makes sure you have enough rows
    fig, ax = plt.subplots(nrows=rows, ncols=columns, figsize = figuresize) # You'll want to specify your figsize
    ax = ax.ravel() # Ravel turns a matrix into a vector, which is easier to iterate
    fig.subplots_adjust(hspace = 0.3, wspace = 0.3)
    for i, column in enumerate(dependentcol): # Gives us an index value to get into all our lists
        try:
            ax[i].scatter(x = data[column], y = independentcol)
            ax[i].set_title(title[i])
            ax[i].set_xlabel(x_label[i])
            ax[i].set_ylabel(y_label[i])
        except: 
            pass
    plt.show()
```

```python
dependents = train.columns.tolist()
```

```python
train['SalePrice']
```

```python
plt.style.use('ggplot')
```

```python
subplot_scat(train, train['SalePrice'],dependents, title=dependents, figuresize = (30,160), columns = 3)
```

```python
def subplot_seascat(data, independentcol, dependentcol, title=None, x_label=None, y_label=None,figuresize = None, columns = 2):
    rows = int(round((len(dependentcol)/columns),0)) # Makes sure you have enough rows
    fig, ax = plt.subplots(nrows=rows, ncols=columns, figsize = figuresize) # You'll want to specify your figsize
    ax = ax.ravel() # Ravel turns a matrix into a vector, which is easier to iterate
    fig.subplots_adjust(hspace = 0.3, wspace = 0.3)
    for i, column in enumerate(dependentcol): # Gives us an index value to get into all our lists
        try:
            sns.scatterplot(x = data[column], y = independentcol, ax=ax[i])
            ax[i].set_title(title[i])
            ax[i].set_xlabel(x_label[i])
            ax[i].set_ylabel(y_label[i])
        except: 
            pass
    plt.show()
```

```python
subplot_seascat(train, train['SalePrice'],dependents, title=dependents, figuresize = (30,160), columns = 3)
```

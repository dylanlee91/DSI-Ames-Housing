## Modeling

Four different models were tested using cross validation on training data:

- ordinary linear regression
- ridge regression
- lasso regression
- elastic net regression

Prior to modeling, we will do a scaling of our feature variables so that they have a mean of 0 and standard deviation of 1. This helps to avoid the case where one or several features dominate others in magnitude, and as a result, the model hardly picks up the contribution of the smaller scale variables, even if they are strong.




These four models were tessted X times as outlined below:

1. Round 1: Test using all the original feature variables
- Round 2: Test using all the unique features that were retained by Lasso and ElasticNet in Round 1

- Round 3: Perform RFE using Ridge to obtain 25 top predictors and test all four models using them

- Round 4: Perform RFE using ElasticNet to obtain 25 top predictors and test all four models using them


### Set up helper functions for Linear Regression, Lasso, Ridge and ElasticNet

```python
def linear_reg(X_train, X_test, y_train, y_test):
    lr = LinearRegression()   
    lr.fit(X_train, y_train)
    y_train_pred = lr.predict(X_train)
    y_test_pred = lr.predict(X_test)
    
    training_R2 = lr.score(X_train, y_train) 
    training_RMSE = np.sqrt(mean_squared_error(y_train, y_train_pred))
    
    testing_R2 = lr.score(X_test, y_test) 
    testing_RMSE = np.sqrt(mean_squared_error(y_test, y_test_pred))    
      
    # k-fold cross validation score
    mean_mse_cv = np.abs(cross_val_score(lr, X_train, y_train, cv=5, scoring='neg_mean_squared_error').mean())
    mean_r2_cv = cross_val_score(lr, X_train, y_train, cv=10).mean()
    
    print('---------- LINEAR REGRESSION -----------')
    print('')
    print(f'Training R2 score: {training_R2:.5f}')
    print(f'Validation R2 score: {testing_R2:.5f}')
    print(f'Cross validation mean R2: {mean_mse_cv:.5f}')
    print('')
    print(f'Training RMSE: {training_RMSE:.5f}')
    print(f'Validation RMSE: {testing_RMSE:.5f}')
    print(f'Cross validation mean RMSE: {np.sqrt(mean_mse_cv):.5f}')  
    print('----------------------------------------')    
        
    # plot predictions
    f = plt.figure(figsize=(15,6))
    ax1 = f.add_subplot(121)
    ax2 = f.add_subplot(122)
    ax1.scatter(y_train_pred, y_train, c='limegreen', s=10, marker = "s", label="Training data")
    ax1.scatter(y_test_pred, y_test, c='white', s=12, marker = "s", edgecolor='midnightblue', linewidths=1, label='Validation data')   
    ax1.set_title('Actual vs. Predicted Values', fontsize=16)
    ax1.set_xlabel('Predicted values', fontsize=15)
    ax1.set_ylabel('Actual values', fontsize=15)
    ax1.legend(loc = 0)
    ax1.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], c='red')
    
    # plot residuals
    ax2.scatter(y_train_pred, y_train_pred - y_train, c='limegreen', s=10, marker = "s", label="Training data")
    ax2.scatter(y_test_pred, y_test_pred - y_test, c='white', s=12, marker = "s", edgecolor='midnightblue', linewidths=1, label='Validation data')
    ax2.set_title("Residuals", fontsize=16)
    ax2.set_xlabel("Predicted values", fontsize=15)
    ax2.set_ylabel("Residuals", fontsize=15)
    ax2.legend(loc=0)
    ax2.hlines(y=0, xmin = y_test.min(), xmax = y_test.max(), color = "blue")
    
    # Coefficients
#     coef_dict = pd.DataFrame({'feature': X.columns,
#                  'coefficient': lr.coef_})    
#     print(coef_dict.sort_values('coefficient', ascending=False))
```

```python
def ridge_reg(X_train, X_test, y_train, y_test):
    r_alphas = np.logspace(0, 5, 200)
    rr = RidgeCV(alphas=r_alphas, cv=5)
    rr.fit(X_train, y_train)
    
    optimal_alpha = rr.alpha_    
    
    # take optimal alpha from CV search and create a new model with the optimal alpha and fit on the
    # entire train dataset
    optimal_ridge = Ridge(alpha=optimal_alpha, random_state=42)
    
    optimal_ridge.fit(X_train,y_train)
    y_test_pred = optimal_ridge.predict(X_test)
    y_train_pred = optimal_ridge.predict(X_train)

    training_R2 = optimal_ridge.score(X_train, y_train) 
    training_RMSE = np.sqrt(mean_squared_error(y_train, y_train_pred))

    testing_R2 = optimal_ridge.score(X_test, y_test)
    testing_RMSE = np.sqrt(mean_squared_error(y_test, y_test_pred))
    
    optimal_ridge_mean_R2 = cross_val_score(optimal_ridge, X_train, y_train, cv=5).mean() # R2
    optimal_ridge_cv_mean_mse = np.abs((-cross_val_score(optimal_ridge, X_train, y_train, cv=5, 
                                                  scoring='neg_mean_squared_error')).mean())
 
    print('---------- RIDGE REGRESSION ----------')
    print('')
    print(f'Optimal alpha: {optimal_alpha:.10f}')
    print('')
    print(f'Training R2 score: {training_R2:.5f}')
    print(f'Validation R2 score: {testing_R2:.5f}')
    print(f'Cross validation mean R2: {optimal_ridge_mean_R2:.5f}')
    print('')
    print(f'Training RMSE: {training_RMSE:.5f}')
    print(f'Validation RMSE: {testing_RMSE:.5f}')
    print(f'Cross validation mean RMSE: {np.sqrt(optimal_ridge_cv_mean_mse):.5f}') 
    print('--------------------------------------')

    # plot predictions
    f = plt.figure(figsize=(15,6))
    ax1 = f.add_subplot(121)
    ax2 = f.add_subplot(122)
    ax1.scatter(y_train_pred, y_train, c='limegreen', s=10, marker = "s", label="Training data")
    ax1.scatter(y_test_pred, y_test, c='white', s=12, marker = "s", edgecolor='midnightblue', linewidths=1, label='Validation data')   
    ax1.set_title('Actual vs. Predicted Values', fontsize=16)
    ax1.set_xlabel('Predicted values', fontsize=15)
    ax1.set_ylabel('Actual values', fontsize=15)
    ax1.legend(loc = 0)
    ax1.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], c='red')
    
    # plot residuals
    ax2.scatter(y_train_pred, y_train_pred - y_train, c='limegreen', s=10, marker = "s", label="Training data")
    ax2.scatter(y_test_pred, y_test_pred - y_test, c='white', s=12, marker = "s", edgecolor='midnightblue', linewidths=1, label='Validation data')
    ax2.set_title("Residuals", fontsize=16)
    ax2.set_xlabel("Predicted values", fontsize=15)
    ax2.set_ylabel("Residuals", fontsize=15)
    ax2.legend(loc=0)
    ax2.hlines(y=0, xmin = y_test.min(), xmax = y_test.max(), color = "blue")
    
    # Coefficients
    coef_dict = pd.DataFrame({'feature': X.columns,
                 'ridge_coef': optimal_ridge.coef_,
                 'abs_ridge_coef': np.abs(optimal_ridge.coef_)}).sort_values('abs_ridge_coef', ascending=False)    
    return coef_dict
```

```python
def lasso_reg(X_train, X_test, y_train, y_test):
    lasso_model = LassoCV(n_alphas=100, cv=5, random_state=42)   
    lasso_model.fit(X_train, y_train)
    
    optimal_alpha = lasso_model.alpha_
    
    optimal_lasso = Lasso(alpha=optimal_alpha, max_iter=10_000, random_state=42)
    optimal_lasso.fit(X_train, y_train)
    y_test_pred = optimal_lasso.predict(X_test)
    y_train_pred = optimal_lasso.predict(X_train)    
    
    training_R2 = optimal_lasso.score(X_train, y_train) 
    training_RMSE = np.sqrt(mean_squared_error(y_train, y_train_pred))
    
    testing_R2 = optimal_lasso.score(X_test, y_test)
    testing_RMSE = np.sqrt(mean_squared_error(y_test, y_test_pred))
    
    optimal_lasso_mean_R2 = cross_val_score(optimal_lasso, X_train, y_train, cv=5).mean() # R2
    optimal_lasso_cv_mean_mse = np.abs((-cross_val_score(optimal_lasso, X_train, y_train, cv=5, 
                                                  scoring='neg_mean_squared_error')).mean())    
    
    print('---------- LASSO REGRESSION ----------')
    print('')
    print(f'Optimal alpha: {optimal_alpha:.10f}')
    print('')
    print(f'Training R2 score: {training_R2:.5f}')
    print(f'Validation R2 score: {testing_R2:.5f}')
    print(f'Cross validation mean R2: {optimal_lasso_mean_R2:.5f}')
    print('')
    print(f'Training RMSE: {training_RMSE:.5f}')
    print(f'Validation RMSE: {testing_RMSE:.5f}')
    print(f'Cross validation mean RMSE: {np.sqrt(optimal_lasso_cv_mean_mse):.5f}')   
    print('--------------------------------------')
    
    # plot predictions
    f = plt.figure(figsize=(15,6))
    ax1 = f.add_subplot(121)
    ax2 = f.add_subplot(122)
    ax1.scatter(y_train_pred, y_train, c='limegreen', s=10, marker = "s", label="Training data")
    ax1.scatter(y_test_pred, y_test, c='white', s=12, marker = "s", edgecolor='midnightblue', linewidths=1, label='Validation data')   
    ax1.set_title('Actual vs. Predicted Values', fontsize=16)
    ax1.set_xlabel('Predicted values', fontsize=15)
    ax1.set_ylabel('Actual values', fontsize=15)
    ax1.legend(loc = 0)
    ax1.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], c='red')
    
    # plot residuals
    ax2.scatter(y_train_pred, y_train_pred - y_train, c='limegreen', s=10, marker = "s", label="Training data")
    ax2.scatter(y_test_pred, y_test_pred - y_test, c='white', s=12, marker = "s", edgecolor='midnightblue', linewidths=1, label='Validation data')
    ax2.set_title("Residuals", fontsize=16)
    ax2.set_xlabel("Predicted values", fontsize=15)
    ax2.set_ylabel("Residuals", fontsize=15)
    ax2.legend(loc=0)
    ax2.hlines(y=0, xmin = y_test.min(), xmax = y_test.max(), color = "blue")
    
    # Coefficients
    coef_dict = pd.DataFrame({'feature': X.columns,
                 'lasso_coef': optimal_lasso.coef_,
                 'abs_lasso_coef': np.abs(optimal_lasso.coef_)}).sort_values('abs_lasso_coef', ascending=False)    
    return coef_dict
```

```python
def elastic_net(X_train, X_test, y_train, y_test):
    enet_model = ElasticNetCV(n_alphas=100, l1_ratio=np.linspace(0.01, 1.0, 25), cv=5, random_state=42)
    enet_model = enet_model.fit(X_train, y_train)

    optimal_alpha = enet_model.alpha_
    optimal_l1ratio = enet_model.l1_ratio_
    # If l1_ratio = 1, the penalty would be L1 penalty. If l1_ratio = 0, the penalty would be an L2 penalty. 
    # If the value of l1 ratio is between 0 and 1, the penalty would be the combination of L1 and L2.
      
    optimal_enet = ElasticNet(alpha = optimal_alpha, l1_ratio = optimal_l1ratio, max_iter=10_000)
    optimal_enet.fit(X_train, y_train)
    
    # generate predictions for both test and train data
    y_test_pred = optimal_enet.predict(X_test)
    y_train_pred = optimal_enet.predict(X_train)    
    
    training_R2 = optimal_enet.score(X_train, y_train) 
    training_RMSE = np.sqrt(mean_squared_error(y_train, y_train_pred))
    
    # evaluate the model using validation data
    testing_R2 = optimal_enet.score(X_test, y_test)
    testing_RMSE = np.sqrt(mean_squared_error(y_test, y_test_pred))
    
    optimal_enet_mean_R2 = cross_val_score(optimal_enet, X_train, y_train, cv=5).mean() # R2
    optimal_enet_cv_mean_mse = np.abs((-cross_val_score(optimal_enet, X_train, y_train, cv=5, 
                                                  scoring='neg_mean_squared_error')).mean())      

    print('------------ ELASTIC NET ------------')
    print('')
    print(f'Optimal alpha: {optimal_alpha:.10f}')
    print(f'Optimal L1 ratio: {optimal_l1ratio:.5f}')
    print('')
    print(f'Training R2 score: {training_R2:.5f}')
    print(f'Validation R2 score: {testing_R2:.5f}')
    print(f'Cross validation mean R2: {optimal_enet_mean_R2:.5f}')
    print('')
    print(f'Training RMSE: {training_RMSE:.5f}')
    print(f'Validation RMSE: {testing_RMSE:.5f}')
    print(f'Cross validation mean RMSE: {np.sqrt(optimal_enet_cv_mean_mse):.5f}')
    print('--------------------------------------')
    
    # plot predictions
    f = plt.figure(figsize=(15,6))
    ax1 = f.add_subplot(121)
    ax2 = f.add_subplot(122)
    ax1.scatter(y_train_pred, y_train, c='limegreen', s=10, marker = "s", label="Training data")
    ax1.scatter(y_test_pred, y_test, c='white', s=12, marker = "s", edgecolor='midnightblue', linewidths=1, label='Validation data')   
    ax1.set_title('Actual vs. Predicted Values', fontsize=16)
    ax1.set_xlabel('Predicted values', fontsize=15)
    ax1.set_ylabel('Actual values', fontsize=15)
    ax1.legend(loc = 0)
    ax1.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], c='red')
    
    # plot residuals
    ax2.scatter(y_train_pred, y_train_pred - y_train, c='limegreen', s=10, marker = "s", label="Training data")
    ax2.scatter(y_test_pred, y_test_pred - y_test, c='white', s=12, marker = "s", edgecolor='midnightblue', linewidths=1, label='Validation data')
    ax2.set_title("Residuals", fontsize=16)
    ax2.set_xlabel("Predicted values", fontsize=15)
    ax2.set_ylabel("Residuals", fontsize=15)
    ax2.legend(loc=0)
    ax2.hlines(y=0, xmin = y_test.min(), xmax = y_test.max(), color = "blue")
    
    # Coefficients
    coef_dict = pd.DataFrame({'feature': X.columns,
                 'enet_coef': optimal_enet.coef_,
                 'abs_enet_coef': np.abs(optimal_enet.coef_)}).sort_values('abs_enet_coef', ascending=False)    
    return coef_dict
```

## Establish baseline score

For my baseline model, I will use DummyRegressor to create a dummy regression that always predicts the mean value of the target variable.

```python
X = train_dummy.drop(columns=['Id', 'PID', 'SalePrice'])
y = train_dummy['SalePrice']

# make test and training split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# create a dummy regressor
dummy_reg = DummyRegressor(strategy='mean')

# scale X data
ss = StandardScaler()
ss.fit(X_train)

X_train_scaled = ss.transform(X_train)
X_test_scaled = ss.transform(X_test)

# "train" dummy regressor
dummy_reg.fit(X_train_scaled, y_train)
dummy_reg.score(X_test_scaled, y_test)
```

Our baseline model's R2 is close to 0, which indicates that the model explains none of the variability of `SalePrice` around its mean.

```python
linear_reg(X_train_scaled, X_test_scaled, y_train, y_test)
```

**Observation**: The R2 on the validation data and the mean CV R2 both have extreme values. The linear regression model visibly performed poorly, with several extreme outlier values predicted. This implies that there is serious overfitting in the Linear Regression model with 169 feature variables.

**What does it mean to have a negative $R^2$?**

Remember that $R^2$ is 1 minus the error of your regression model divided by the error of the baseline model. A negative $R^2$ means that the regression model is performing *worse* than the baseline model. In the context of fitting our data on one sample of data and scoring on another sample, this means that we would have been better off making predictions on the test sample just using the mean of the target variable in our training set.

- To further reduce the number of predictors, let's now try an automated feature selection method called Recursive Feature Elimination (RFE). 


- We will conduct RFE thrice with Lasso, Ridge, and ElasticNet as estimators separately, and see which set of features produces the best results.

## Recursive Feature Elimination (using Ridge as estimator)

```python
# Same X and y as attempt 2
X = train_dummy.drop(columns=['Id', 'PID', 'SalePrice'] + lasso1_enet1_dropped)
y = train_dummy['SalePrice']

ss = StandardScaler()
ss.fit(X)

X_scaled = ss.transform(X)

r_alphas = np.logspace(0, 5, 200)
ridge_model = RidgeCV(alphas=r_alphas, cv=5)
rfe_ridge = RFE(estimator=ridge_model, step=1, n_features_to_select=25)
rfe_ridge.fit(X_scaled, y)
```

```python
rfe_ridge.ranking_
```

```python
# list of the top 25 features selected by RFE using Ridge
top25_ridge_rfe = X.loc[:, rfe_ridge.support_].columns
```

```python
# this returns the coefficients for the top 25 features
print ('coefficients', rfe_ridge.estimator_.coef_)
```

```python
# save the top 25 features and their coefficients into a dataframe
df_coef_ridge = pd.DataFrame({'feature': top25_ridge_rfe,
              'coefficient': np.transpose(rfe_ridge.estimator_.coef_),
              'abs_coefficient': np.abs(np.transpose(rfe_ridge.estimator_.coef_))}).\
                sort_values('abs_coefficient', ascending=False)
```

```python
df_coef_ridge
```

```python
# plot the predictors from RFE with Ridge
plt.figure(figsize=(12,7))
sns.barplot(data=df_coef_ridge, x='coefficient', y='feature', palette="plasma")
plt.title('Top 25 features from RFE using Ridge', fontsize=16)
plt.xlabel('coefficient', fontsize=14)
plt.ylabel(None)
plt.yticks(fontsize=12)
plt.tight_layout()
```

## Recursive Feature Elimination (using Lasso as estimator)

```python
# Same X and y as attempt 2
X = train_dummy.drop(columns=['Id', 'PID', 'SalePrice'] + lasso1_enet1_dropped)
y = train_dummy['SalePrice']

ss = StandardScaler()
ss.fit(X)

X_scaled = ss.transform(X)

lasso_model = LassoCV(n_alphas=100, cv=5, random_state=42)
rfe_lasso = RFE(estimator=lasso_model, step=1, n_features_to_select=25)
rfe_lasso.fit(X_scaled,y)
```

```python
# list of the top 25 features selected by RFE using Lasso
top25_lasso_rfe = X.loc[:, rfe_lasso.support_].columns
```

```python
# save the top 25 features and their coefficients into a dataframe
df_coef_lasso = pd.DataFrame({'feature': top25_lasso_rfe,
              'coefficient': np.transpose(rfe_lasso.estimator_.coef_),
              'abs_coefficient': np.abs(np.transpose(rfe_lasso.estimator_.coef_))}).sort_values('abs_coefficient', ascending=False)
```

```python
df_coef_lasso
```

## Recursive Feature Elimination (using ElasticNet as estimator)

## Testing on test data

```python
def ridge_test(X, y, test_data):
    ridge_alphas = np.logspace(0, 5, 200)
    ridge = RidgeCV(alphas=ridge_alphas, cv=10)
    ridge.fit(X, y)
    print('Optimal alpha: ', ridge.alpha_)
    
    optimal_ridge = Ridge(alpha=ridge.alpha_)
    optimal_ridge.fit(X, y)
    
    # reverse the log transformation of SalePrice 
    predictions = np.exp(optimal_ridge.predict(test_data))
    
    results = pd.concat([test_dummy[['Id']], pd.DataFrame(predictions,columns=['SalePrice'])], axis=1) 
    
    return results
```

## Limitations

- **Short timeframe**: Our data comprised only transactions between 2006–2010. This is a pretty short timeframe, which makes it hard to capture actual trends in sale prices of houses. Also, new trends may have emerged in the housing market since then, which reduces the usefulness of our analysis. 


- **Missing values**: There were 


- **Other external factors not taken into account**: The value of a house is not purely determined by the features we have analysed. There are many other factors that influence the price of a house, such as the current economic climate, housing policies, proximity of the house to amenities like malls and schools. 


- **Limited generalisability to houses outside of Ames, Iowa**: The predictive model we've built only applies to the houses in Ames and will not be accurate when applied to houses in other cities or countries.

# 2. Imputation of missing values

As discussed in the previous section, many null values can be added in by deductive imputation.

```python
# First, we drop all the problematic columns and rows identified
# Dropping columns
df.drop(columns =['lotfrontage', 'poolarea', 'poolqc'],inplace=True)
# Dropping 
df.drop([1712],inplace=True)
df.head()
```

```python
def autofiller(data):
    data.fillna(0, inplace=True)
    for col in data.columns:
        if data[col].dtypes == 'object':
            data[col].replace({0:'None'},inplace=True)
        else:
            pass
    return data     
```

```python
# Imputing the missing values with either none or 0
autofiller(df)
df.head()
```

Of the variables not identified as having little correlation, the following sets of variables have identified as likely to subsets of each other or otherwise multicollinear in some other way.

7. garageyearbuilt and the other age metrics: given that it is highly likely that the garage and house were built in the same year, only the age metrics will be used
8. newageatsale and ageatsale: Given the large amount of overlap between the two sets of data, the newness of the house and its effect on the sale price will be used (newageatsale).

exterqual, bsmtqual and garagequal

# 5.1 Feature Engineering dummies for categories

```python
# As per the discussion, lists of the two types of categories based on deduction are created
usefulcat = ['mszoning','street','landcontour','neighborhood','housestyle',
                   'roofstyle','roofmatl','masvnrtype','foundation','bsmtexposure',
                   'functional', 'fireplacequ', 'garagetype','garagefinish']
unusefulcat = ['alley','lotshape','utilities','landslope','lotconfig','bldgtype', 
                     'heating', 'centralair','electrical', 'paveddrive', 'fence', 'miscfeature']
```

```python
df = pd.get_dummies(df,columns=usefulcat)
df.head()
```

```python
def twopartdummies(column1, column2, prefix = ''):
    dum = df[column1].str.get_dummies().add(df[column2].str.get_dummies(),fill_value=0)
    for col in dum.columns:
        dum[col] = dum[col].map({2:1, 1:1,0:0}, na_action='ignore')
    dum = dum.astype(int).add_prefix(prefix)
    return dum    
```

```python
# Generating dummies for condition 1 and 2
df = pd.concat([df,twopartdummies('condition1', 'condition2', prefix = 'condition_')],axis=1)
df.head()
```

```python
# Generating dummies for basement finishes type 1 and 2
df = pd.concat([df,twopartdummies('bsmtfintype1', 'bsmtfintype2', prefix = 'basement_')],axis=1)
df.head()
```

```python
df = pd.concat([df,twopartdummies('exterior1st', 'exterior2nd', prefix = 'exterior_')],axis=1)
df.head()
```
